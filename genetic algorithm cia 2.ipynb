```python
import pandas as pd
import numpy as np
import pygad.kerasga as pg
import pygad
import tensorflow.keras
import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.losses import BinaryCrossentropy
```

```python
df=pd.read_csv("Bank_Personal_Loan_Modelling.csv")
```

```python
df.drop(["ID", "ZIP Code"],axis=1,inplace=True)
```

```python
df = df[['Age', 'Experience', 'Income', 'Family', 'CCAvg','Education', 'Mortgage', 'Securities Account','CD Account', 'Online', 'CreditCard', 'Personal Loan']]
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
model= keras.Sequential(
    [
     keras.layers.Dense(3,activation="relu",input_shape=[11]),
     keras.layers.Dense(3,input_dim=11,activation="relu"),
     keras.layers.Dense(1,activation="sigmoid")
    ]
)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=69)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
gg= pg.KerasGA(model=model,num_solutions=20)
```

```python
def fitness_func(solution, sol_idx):
    global x, y, keras_ga, model
    model_weights_matrix = pygad.kerasga.model_weights_as_matrix(model=model,weights_vector=solution)                                                                                                                  
    model.set_weights(weights=model_weights_matrix)
    predictions = model.predict(x)
    p = tensorflow.keras.losses.BinaryCrossentropy()
    solution_fitness = 1.0 / (p(y, predictions).numpy() + 0.00000001)
    return solution_fitness

def callback_generation(ga):
    print("Generation = {generation}".format(generation=ga.generations_completed))
    print("Fitness    = {fitness}".format(fitness=ga.best_solution()[1]))
```

```python
ga = pygad.GA(num_generations=10, num_parents_mating=7, 
                       initial_population=gg.population_weights,
                       fitness_func=fitness_func,
                       on_generation=callback_generation)
ga.run()
```
